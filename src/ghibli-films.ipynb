```python
#Import das libs utilizadas
import matplotlib.pyplot as plt
import pandas as pd
import requests
```

```python
url = 'https://ghibliapi.herokuapp.com/films'
```

```python
#Chamada da API 
r = requests.get(url)
```

```python
r
```

```python
#Extrai os dados .json
json = r.json()
```

```python
json
```

```python
#Cria o DataFrame 
df = pd.DataFrame(json)
```

```python
df
```

```python
df.columns = ['id', 'titulo', 'titulo_japones', 'titulo_romanizado',
       'descricao', 'diretor', 'produtor', 'data_lancamento', 'duracao',
       'nota_avaliacao', 'pessoas', 'especies', 'ambientes', 'veiculos', 'url']
```
